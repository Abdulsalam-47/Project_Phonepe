```python
#Explanation purpose
#Created after loading json data as now we know what are the things we need to get
agg_insur_column={"States":[], "Years":[], "Quarter":[], "Transaction_type":[], "Transaction_count":[], "Transaction_amount":[]}
#this loop used to open each state in loop 1 after the other 
for state in agg_insur_list:
    cur_states=path_agg_insur+state+"/"
    #print(cur_states)
    #agg_year_list used because if we open state folder we get year folders
    agg_year_list=os.listdir(cur_states)
    #print(agg_year_list)
    for year in agg_year_list:
        cur_year=cur_states+year+"/"
        #print(cur_year)
        agg_file_list=os.listdir(cur_year)
        #print(agg_file_list)
        #to open the files one by one
        for file in agg_file_list:
            cur_file=cur_year+file
            #print(cur_file)
            #to read the file
            data=open(cur_file,"r")
            
            A=json.load(data)
            #print(A)
            #for loop to get data from transactionData from the json files
            for i in A["data"]["transactionData"]:
                #print(i)
                name=i["name"]
                #print(name)
                count=i["paymentInstruments"][0]["count"]
                #print(count)
                amount=i["paymentInstruments"][0]["amount"]
                #print(amount)
                agg_insur_column["Transaction_type"].append(name)
                agg_insur_column["Transaction_count"].append(count)
                agg_insur_column["Transaction_amount"].append(amount)
                agg_insur_column["States"].append(state)
                agg_insur_column["Years"].append(year)
                agg_insur_column["Quarter"].append(int(file.strip(".json")))
    
    #converting to DataFrame
    aggre_insurance=pd.DataFrame(agg_insur_column)
    print(aggre_insurance)

    aggre_insurance["States"]= aggre_insurance["States"].str.replace('andaman-&-nicobar-islands', 'Andaman & Nicobar')#name changed
    aggre_insurance["States"]=aggre_insurance["States"].str.replace('-',' ')# remove hyphen and replaced it with space in states names
    aggre_insurance["States"]=aggre_insurance["States"].str.title()#Changing the states name starting letter to caps
    aggre_insurance['States']=aggre_insurance["States"].str.replace('Dadra & Nagar Haveli & Daman & Diu', 'Dadra and Nagar Haveli and Daman and Diu')
```

```python
#aggre_insurance

path_agg_insur="D:/Guvi Projects/Phonepe/pulse/data/aggregated/insurance/country/india/state/"
agg_insur_list=os.listdir(path_agg_insur)

agg_insur_column={"States":[], "Years":[], "Quarter":[], "Transaction_type":[], "Transaction_count":[], "Transaction_amount":[]}

#Iterating Over States
#This loop iterates over each entry (state) in the agg_insur_list, which contains the names of directories (or files) within the path_agg_insur directory.
for state in agg_insur_list:
    cur_states=path_agg_insur+state+"/"
    agg_year_list=os.listdir(cur_states)

    #Iterating Over Years
    #This nested loop iterates over each entry (year) in the agg_year_list, which represents directories for each year within the current state directory
    for year in agg_year_list:
        cur_year=cur_states+year+"/"
        agg_file_list=os.listdir(cur_year)

        #Processing Files
        #This innermost loop iterates over each file in the agg_file_list within the current year directory.
        for file in agg_file_list:
            cur_file=cur_year+file
            data=open(cur_file,"r")

            A=json.load(data)
            for i in A["data"]["transactionData"]:
                name=i["name"]
                count=i["paymentInstruments"][0]["count"]
                amount=i["paymentInstruments"][0]["amount"]
                agg_insur_column["Transaction_type"].append(name)
                agg_insur_column["Transaction_count"].append(count)
                agg_insur_column["Transaction_amount"].append(amount)
                agg_insur_column["States"].append(state)
                agg_insur_column["Years"].append(year)
                agg_insur_column["Quarter"].append(int(file.strip(".json")))

aggre_insurance=pd.DataFrame(agg_insur_column)

aggre_insurance["States"]=aggre_insurance["States"].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
aggre_insurance["States"]=aggre_insurance["States"].str.replace("-"," ")
aggre_insurance["States"]=aggre_insurance["States"].str.title()
aggre_insurance["States"]=aggre_insurance["States"].str.replace('Dadra & Nagar Haveli & Daman & Diu', 'Dadra and Nagar Haveli and Daman and Diu')

#aggre_insurance Table Creation

#aggregated_insurance
create_query1=create_query1 = '''
                                CREATE TABLE IF NOT EXISTS aggregated_insurance(
                                    States VARCHAR(255),
                                    Years INT,
                                    Quarter INT,
                                    Transaction_type VARCHAR(255),
                                    Transaction_count BIGINT,
                                    Transaction_amount BIGINT,
                                    PRIMARY KEY (States, Years, Quarter, Transaction_type)
                                )
                                '''

cursor.execute(create_query1)
mydb.commit()

insert_query1='''INSERT IGNORE INTO aggregated_insurance(States, Years, Quarter, Transaction_type, 
                                                    Transaction_count,
                                                    Transaction_amount)
                                                    
                                                    values(%s,%s,%s,%s,%s,%s)'''
data=aggre_insurance.values.tolist()
cursor.executemany(insert_query1,data)
mydb.commit()
```
